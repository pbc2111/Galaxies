```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

calzettiFilename = 'Calzetti01'

calzettiAttenuation = np.zeros((3813,2))

with open(calzettiFilename) as fp:
    for i, line in enumerate(fp):
        if i != 0:
            data = line.split()
            calzettiAttenuation[i-1, 0] = data[0]
            calzettiAttenuation[i-1, 1] = data[1]

#convert to microns from angstroms
calzettiAttenuation[:,0] *= 1e-4

#E(B − V ) = 0.1 and E(B − V ) = 1.0
A_lower = calzettiAttenuation[:,1] * 0.1

#plot Attenuation curves
fig = plt.figure()
plt.style.use('ggplot')
plt.xlabel(r'\lambda')
plt.ylabel(r'A_{\lambda}')
plt.loglog(calzettiAttenuation[:,0], A_lower)
plt.loglog(calzettiAttenuation[:,0], calzettiAttenuation[:,1])
plt.tight_layout()
plt.show()

fig.savefig("Attenuation_lower.pdf", bbox_inches='tight')
```
